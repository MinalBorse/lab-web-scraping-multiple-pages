```python
#initialize empty lists
title = []
artist = []


# define the number of iterations of our for loop
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select("cite.title"))

# iterate through the result set and retrive all the data
for i in range(num_iter):
    title.append(soup.select("cite.title")[i].get_text())
    artist.append(soup.select("em.artist")[i].get_text())
    

print(title)
print(artist)

# each list becomes a column
songs = pd.DataFrame({"title":title,
                       "artist":artist
                      })
```

```python
#Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page: url ='https://en.wikipedia.org/wiki/Python'
html = urlopen("https://en.wikipedia.org/wiki/Python")
list_python = BeautifulSoup(html)
for link in list_python.findAll("a"):
  if 'href' in link.attrs:
    print(link.attrs['href'])
```

```python
#Display the top 10 languages by number of native speakers stored in a pandas dataframe: url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
html = urlopen("https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers")
top_10 = BeautifulSoup(html)
for link in top_10.findAll("a"):
  if 'href' in link.attrs:
    print(link.attrs['href'])
```
